```python
###AULA 3

##Importar bibliotecas e carregar dados

import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
dados.columns = dados.columns.str.replace("-", "")
```

```python
#Plotar tabela cruzada de observações

pd.crosstab(dados['dose'], dados['tempo'])
```

```python
#Plotar tabela cruzada de observações com normalização


(pd.crosstab(dados['tempo'], dados['dose'], normalize=True))
```

```python
#Plotar tabela cruzada de observações com normalização adicionando o fator tratamento

pd.crosstab([dados['tempo'], dados['dose']], dados['tratamento'],)
```

```python
#É possível selecionar dados com o argumento values e realizar operações como 
#média da variável para cada grupo

pd.crosstab([dados['tempo'], dados['dose']], dados['tratamento'], 
            normalize=False, values=dados['g0'], aggfunc='mean')
```

```python
##Gráfico de dispersão

sns.scatterplot(x='g0', y='g3', data=dados)

plt.show()
```

```python
###Gráfico de dispersão com as variáveis g0 e g8 para observar relação dos dados


sns.scatterplot(x='g0', y = 'g8', data=dados)
```

```python
#Usar sns.lmplot para plotar uma linha que representa o modelo linear
#parametro line_kws para mudar a cor

sns.lmplot(x='g0', y='g8', data=dados, line_kws={'color':'red'})

plt.show()
```

```python
#É possível produzir gráficos para cada combinação de tratamentos
#No caso em cada coluna vai representar com ou sem droga
#Cada linha um intervalo de tempo de tratamento

sns.lmplot(x='g0', y='g8', data=dados, line_kws={'color':'red'}, col='tratamento', row='tempo')

plt.show()
```

```python
#Calcular a correlação entre as variáveis

dados.loc[:,'g0':'g771'].corr()
```

```python
corr = dados.loc[:,'g0':'g50'].corr()
```

```python
#Mapa de calor

corr = dados.loc[:,'g0':'g50'].corr()

import numpy as np


# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.8})
```

```python
#Mapa de calor para C

corr_celular = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .8})

```
